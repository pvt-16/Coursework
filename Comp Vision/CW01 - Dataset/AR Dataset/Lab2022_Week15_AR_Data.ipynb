## Read NPY File for AR Dataset

```
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt # plotting
import numpy as np # linear algebra
import os # accessing directory structure
import cv2 

## put the AR npz file in the root folder /Kaggle
ar_database = np.load("./Kaggle/ar_database_stacked_128x128.npz")["stack"]

fig = plt.figure(dpi = 300)

for i in range(0, 18):
    ax = plt.subplot(6, 9, i + 1)
    plt.gray()
    ax.imshow(ar_database[i, :, :], interpolation='nearest')
    ax.set_axis_off()

fig.subplots_adjust(hspace=0.01)

for i in range(0,ar_database.shape[0]):
  cv2.imwrite('./Kaggle/'+f'{i:04}'+'.jpg', ar_database[i, :, :])

!zip -r /content/file.zip /content/Kaggle
```
