```python
class Tree:  #class definition
    def __init__(self, content, left=None, right=None):  
        self.content = content
        self.left = left
        self.right = right

def averagePositive(tree):  #function to calculate the average
    [sumPos, countPos] = _avgPos(tree, 0) #calls private method
    return round(sumPos/ (countPos if countPos>0 else 1), 5) 

#Private method to calculate average
#it returns a tuple of current node value and a counter that tracks all positive numbers encountered
def _avgPos(tree, countPos) -> tuple():
    if tree is None:  #Leaf node
        return (0, countPos)
    else:
        localSum = 0
        #Calculate Left side
        localTree =  _avgPos(tree.left, countPos) 
        #Assign local sum of left branch and counter value
        localSum += localTree[0] 
        countPos = localTree[1]   #Returns the counter value
        # If central node is a positive number, it is added 
        if tree.content >=0:
            localSum += int(tree.content)
            countPos +=1
        #Calculate the Right side
        localTree = _avgPos(tree.right, countPos)
        #Add evaluated values from the right branch 
        localSum += localTree[0]  
        countPos = localTree[1] #Returns the counter value
        return (localSum,countPos)
```

```python
class Tree:
    # defining the constructor
    def __init__(self, content, left = None, right = None):
        # assigning the content
        self.content = content
        #assigning the left node
        self.left = left
        #assigning the right node
        self.right = right

# Defining the sum function

def sum_values(tree):

    # If it's none then we return 0
    if tree is None:
        return 0
    # If its positive we do calculations
    else :
        # return the results
        s1 =  (sum_values(tree.right) + sum_values(tree.left) )
        if tree.content >=0:
          s1 += tree.content
        return s1
    # else:
    #     # if its negative return 0
    #     return 0
# defining count function
def count_values(tree):
    # If it's none then we return 0
    if tree is None:
        return 0
    # If its positive we do calculations
    else: 
        # Increase the counter
        c1 = count_values(tree.right) + count_values(tree.left)
        if tree.content >= 0:
          c1 += 1
        return c1
    # else:
    #     # if its negative return 0
    #     return 0
     
    
def averagePositive(tree):

    #Return the calculation
    return (sum_values(tree)/count_values(tree))
```

## Q2. Queue

```python
class Node:   #defining Class Node
    def __init__(self, content=None, next=None):
        self.content = content
        self.next  = next
        self.priority = -1

class PriorityQueue: # Defining Priority Queue
    def __init__(self):
        self.length = 0
        self.head = None

    def is_empty(self):
        return self.length == 0

    def clear(self):  #To clear the queue
        self.length = 0
        self.head = None

    def insertInQueue(self, content, priority):
        node = Node(content)
        #setting node priority
        node.priority = priority
        #If list is empty the new node goes first
        if self.head is None:
            self.head = node
        else:
             #Find the last node in the list
            last = self.head
            while last.next:
                last = last.next
             #Append the new node
            last.next = node
        self.length = self.length + 1

    def removeFromQueue(self):
      # If there are no elements then we return False
        if self.length == 0:
            return False
        #If there is only 1 element, we can ignore the priority check and return the head element
        if self.length ==1:
            content = self.head.content
            self.length = self.length - 1
            return content
        else:
          #declaring 2 nodes to track the previous node
          #newPrevNote will be used to hold the node that is present before the node that needs to be removed
          #It is needed to point to address of the next node after removal of the node
          #prevNode will be used to track and traverse the list
            prevNode = newPrevNode = self.head
            # The node that needs to be removed will be assigned to this. We start with assuming head node has highest priority
            removeNode = self.head
            highestPriority = removeNode.priority
            # we start with the 2nd node since we have the priority of head node
            currentNode = self.head.next
            while currentNode:
                if currentNode.priority > highestPriority:
                    removeNode = currentNode  #Tracking the node with highest priority
                    newPrevNode = prevNode  #Tracking currentNode's previous node 
                    highestPriority= currentNode.priority  
                #continuing traversal
                currentNode  = currentNode.next  
                prevNode = prevNode.next
            
            #When the first node needs to be removed, we need to change the head variable of the Queue
            if removeNode == self.head:
                self.head = self.head.next
            else:
                newPrevNode.next = removeNode.next #previous node points to the new node
            self.length = self.length - 1
            return removeNode.content
```

```python
class Stack : #Original Parent Class - Stack
    #Week 7 solution code is used as it is. No change in code
    def __init__ ( self ):  #constructor
        self.items = 100*[0]
        self.position = 0

    def push (self , item ):   #inseting a new item to Stack
        if ( self.position < 100) : #only 100 elements
            self.items [ self.position ] = item 
            self.position = self.position + 1
            return True 
        else :
            return False 

    def pop ( self ):  #removing an element from Stack
        if ( self.position <= 0):
            return False 
        else :
            self.position = self.position - 1
            return self.items [ self.position ]
    def empty ( self ):  #clearing the Stack
        self.position = 0
    def is_empty ( self ):      #checking if the Stack is empty
        return ( self.position == 0)

class Node :        #Original Class Node
#Week 8 solution code is used as it is. No change in code
    def __init__ (self , content =None , next = None ):
        self.content = content
        self.next = next

class Queue :   #Original Queue Class
    #Week 8 solution code is used as it is. No change in code
    def __init__ ( self ):     #constructor
        self.length = 0
        self.head = None

    def is_empty ( self ):  #Check if queue is the empty
        return self.length == 0

    def clear ( self ): #Clear the queue
        self.length = 0
        self.head = None

    def insert (self , content ):   #Insert a new element to the queue
        node = Node ( content )
        if self.head is None :
            #If list is empty the new node goes first
            self.head = node
        else:
            # Find the last node in the list
            last = self.head
            while last.next :
                last = last.next
            # Append the new node
            last.next = node
        self.length = self.length + 1

    def remove ( self ):    
        #Remove the first element from the queue
        content = self.head.content
        self.head = self.head.next
        self.length = self.length - 1
        return content

#New Class FibonacciStack inherits Stack class 
class FibonacciStack(Stack):
    def __init__(self):
        #Calling the constructor of the parent class
        Stack.__init__(self)
        ## For keeping track of Fibonacci Sequence. Note that these attributes allow us to get the next number without having to re-calculate the whole sequence.
        self.previous = 1
        self.secondPrevious = 0
        self.numRemoved = 0
    # For pop() -> Week 6 solution code is used as it is. No change in code
    # We must multiply by the correct number before we return the item to the user    
    def pop(self):
        #Calling the pop method of the parent class
        #It returns the value for 'content'
        content = super().pop()

        # Here we modify the number
        if (self.numRemoved == 0):
            self.numRemoved = self.numRemoved + 1
            return 0 * content
        
        if (self.numRemoved == 1):
            self.numRemoved = self.numRemoved + 1
            return 1 * content

        #Tracking the fibonacci numbers
        if (self.numRemoved > 1):
            tmp = self.previous + self.secondPrevious
            self.secondPrevious = self.previous
            self.previous = tmp
            return tmp*content  #return the expected value
            
        #returning False as a default value in case the pop function fails
        return False

    def empty(self):
        #Calling the empty method of the parent class
        super().empty()
        #Resetting the values for the fibonacci series
        self.previous = 1
        self.secondPrevious = 0
        self.numRemoved = 0

#New Class PriorityNode. It inherits the Node class
class PriorityNode(Node):
    
    def __init__(self, content=None, priority=None, next=None): #constructor
        #Calling the constructor of the parent class
        Node.__init__(self, content=content, next=next)
        #priority is set separately in this method
        self.priority = priority

#New Class PriorityQueue. It inherits the Queue class
class PriorityQueue(Queue):
    def __init__(self):
        #Calling the constructor of the parent class
        Queue.__init__(self)

    #we need to repeat the insert method since Queue class is initializing the node inside the insert function
    #and we are using PriorityNode type to track content and priority in Priority Queue 
    def insert(self, content, priority):
        node = PriorityNode( content, priority)
        # Logic in Week 8 solution code is used as it is. No change in code
        if self.head is None :
            #If list is empty the new node goes first
            self.head = node
        else:
            # Find the last node in the list
            last = self.head
            while last.next :
                last = last.next
            # Append the new node
            last.next = node
        self.length = self.length + 1
   
    #Need to override the remove method for Priority Queue since we need to remove by priority
    # For remove() -> Week 8 solution code is used as it is. No change in code
    def remove(self):
        currentNode = self.head
        highestPriorityFound = self.head.priority
        
        while (currentNode != None):
            if (currentNode.priority > highestPriorityFound):
                highestPriorityFound = currentNode.priority
            currentNode = currentNode.next

        #Finding the first node in the queue with highestPriorityFound
        currentNode = self.head
        previousNode = None
        found = False

        while (found == False):
            if (currentNode.priority == highestPriorityFound):
                ## I found the node. Hence, I must now remove it from the Queue and break the loop
                content = currentNode.content

                if (previousNode == None):
                    ## If I have no previous node, then I am the head of the Queue
                    self.head = self.head.next
                else:
                    ## If I have previous node, then I must link it to my next neighbor
                    previousNode.next = currentNode.next

                self.length = self.length - 1

                return content

            ## Prepare for next iteration
            previousNode = currentNode
            currentNode = currentNode.next

        #Returning this as default value
        return -999
```
