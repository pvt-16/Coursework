```python
def normalise(data):
    normalisedData = data.copy()
    rows = data.shape[0]
    cols = data.shape[1]
    for j in range(cols):
        maxElement = np.amax(data[:,j])
        minElement = np.amin(data[:,j])
        for i in range(rows):
            normalisedData[i,j]=(data[i,j]-minElement)/(maxElement-minElement)
    return normalisedData
```

##### Standardization

```python
def standard(data):
    standardData = data.copy()
    
    rows = data.shape[0]
    cols = data.shape[1]

    for j in range(cols):
        sigma = np.std(data[:,j])
        mu = np.mean(data[:,j])

        for i in range(rows):
            standardData[i,j] = (data[i,j] - mu)/sigma

    return standardData
```

### Hierarchical Clustering

```python
data_Sample = np.array([[1,2], [2.5, 4.5], [2, 2], [4,1.5], [4, 2.5]])
```

```python
#Function to calculate distance
#Euclidean distance
import numpy as np

def calc_Euc_dist (a1, a2, b1, b2):
  return ((a1- a2)**2 + (b1 - b2)** 2 )** 0.5

def cal_distance(mat):
  rows = mat.shape[0]
  cols = mat.shape[1]
  res_mat = np.zeros((rows, rows))
  for i in range(rows):
    for j in range(rows):
      total = 0
      for c in range(cols):
        total += (mat[i,c] - mat[j,c])** 2
  
      res_mat[i,j] = total ** 0.5
  return res_mat

def dist(p1, p2):
  total = 0
  for c in range(len(p1)):
    total = total +  (p1[c] - p2[c])**2
  return total** 0.5
```

```python
distanceData = cal_distance(data_Sample)
data_Sample.shape
```

This is the distance matrix. Now we will condense it more.
We need only one half of the matrix since the values are same across the diagonal. Reduce memory usage and computations

Using scipy library

```python
import scipy.spatial.distance as sd
import scipy.cluster as sc
condensedDistance = sd.squareform(distanceData)
# the linkage function generates a hierarchical cluster tree, returning the linkage information in a matrix, Z.
Z = sc.hierarchy.linkage( condensedDistance )
```

```python
import matplotlib.pyplot as plt
plt.figure ( figsize =(6 ,4))
sc.hierarchy.dendrogram(Z)
```

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

#Reading wine data
dataRaw = [];
DataFile = open("wine.txt", "r")

while True:
    theline = DataFile.readline()
    if len(theline) == 0:
         break  
    readData = theline.split(",")
    for pos in range(len(readData)):
        readData[pos] = float(readData[pos]);
    dataRaw.append(readData)

DataFile.close()

data = np.array(dataRaw)
distanceData = cal_distance( data)
condensedDistance = sd.squareform ( distanceData )
Z = sc.cluster.hierarchy.linkage(condensedDistance )
#plt.figure (figsize =(300,100))
sc.cluster.hierarchy.dendrogram(Z)
```

```python
sc.cluster.hierarchy.dendrogram(Z, truncate_mode ="level",p =4)
```

```python
import scipy.cluster as sc

standardisedData = standard(reducedData)
centroids, distortion = sc.vq.kmeans(standardisedData, 2)
```

```python
group1= np.array([])
group2= np.array([])

for d in standardisedData:
  if (dist(d, centroids[0,:]) < dist(d, centroids[1,:])):
    if (len(group1) == 0):
      group1 = d
    else:
      group1 = np.vstack((group1, d))
  else:
    if (len(group2) == 0):
      group2 = d
    else:
      group2 = np.vstack((group2, d))
```
